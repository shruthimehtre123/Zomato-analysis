```python
zomato1=zomato.groupby(['Restaurant Name']).agg({'Dining Rating':'mean'}).reset_index()
zomato1
```

```python
zomato2=zomato1.sort_values(['Dining Rating'],ascending=False).fillna(0).head(10)
zomato2
```

```python
plt.bar(x=zomato2['Restaurant Name'],height=zomato2['Dining Rating'])
plt.xticks(rotation='vertical', ha='center')
plt.show()
```

```python
df_11=df.groupby(['City','Cuisine ']).agg({'Dining Votes':'mean'}).reset_index().fillna(0.0)
```

```python
for i in df_11.City.unique():
    st=df_11[df_11['City']==i]
    st=st.sort_values(by=['Dining Votes'],ascending=False).head(4)
    plt.bar(st['Cuisine '],st['Dining Votes'])
    plt.title("XXXXXXXX"+str(st['City'].unique().tolist()[0])+"XXXXXXX")
    plt.show()
```

```python
df_111=df.groupby(['City','Cuisine ']).agg({'Delivery Rating':'mean'}).reset_index().fillna(0.0)
```

```python
for i in df_111.City.unique():
    st=df_111[df_111['City']==i]
    st=st.sort_values(by=['Delivery Rating'],ascending=False).head(4)
    plt.bar(st['Cuisine '],st['Delivery Rating'])
    plt.title("XXXXXXXX"+str(st['City'].unique().tolist()[0])+"XXXXXXX")
    plt.show()
```

```python
df_1112=df.groupby(['City','Item Name']).agg({'Votes':'sum'}).reset_index().fillna(0.0)
```

```python
for i in df_1112.City.unique():
    st=df_1112[df_1112['City']==i]
    st=st.sort_values(by=['Votes'],ascending=False).head(4)
    plt.bar(st['Item Name'],st['Votes'])
    plt.xticks(rotation='vertical', ha='center')

    plt.title("XXXXXXXX"+str(st['City'].unique().tolist()[0])+"XXXXXXX")
    plt.show()
```
